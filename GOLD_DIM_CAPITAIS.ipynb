```python
import pandas as pd
import numpy as np
from pandas_gbq import to_gbq
```

```python
dim_capitais = pd.read_csv("gs://tech3-dados-covid/capitais.csv",encoding="latin1",sep=";")
```

```python
dim_capitais = dim_capitais.rename(columns={
    "Categoria - Tipo ": "ID_Capital",
    "Categoria - Descrição": "Capital"
})
```

```python
dim_capitais["Capital"] = (
    dim_capitais["Capital"]
    .str.replace("Município de ", "", regex=False)
    .str.replace(r"\s*\(.*\)", "", regex=True)
)
```

```python
to_gbq(
    dim_capitais,
    destination_table="tech3_pesquisa_covid.GOLD_DIM_CAPITAL",
    project_id="project-d5e304e6-93c5-44ee-974",
    if_exists="replace"
)
```
