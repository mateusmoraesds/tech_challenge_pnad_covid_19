```python
import pandas as pd
import numpy as np
from pandas_gbq import to_gbq
```

```python
to_gbq(
    dim_uf,
    destination_table="tech3_pesquisa_covid.GOLD_DIM_UF",
    project_id="project-d5e304e6-93c5-44ee-974",
    if_exists="replace"
)
```
